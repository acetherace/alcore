```python
def bimodal_pdf(x, w1=0.5, w2=0.5, m1=0, v1=1, m2=1, v2=1):
    p1 = norm.pdf(x, loc=m1, scale=v1)
    p2 = norm.pdf(x, loc=m2, scale=v2)
    p = w1*p1 + w2*p2
    return p
```

```python
x = np.linspace(-50, 100, 1000)
```

```python
m1=0;  v1=2
m2=15; v2=2
p_scale=6
p = [bimodal_pdf(xi, v1=2, v2=2, m2=15) for xi in x]
q1 = [norm.pdf(xi, loc=(m2-m1)/2, scale=p_scale) for xi in x]
```

```python
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, p, alpha=0.5, color='tab:blue')
plt.fill_between(x, p, alpha=0.5, color='tab:blue')

ax.plot(x, q1, alpha=0.5, color='tab:green')
plt.fill_between(x, q1, alpha=0.5, color='tab:green')

width=20
x_start = (m2-m1)/2 - width
x_end   = (m2-m1)/2 + width
ax.set_xlim(x_start, x_end)
ax.set_ylim(0, 0.25)

ax.axis('off')

plt.show()
```
